```python
import pandas as pd
ipa15_1 = pd.read_csv("prov_2/ipa_prov/ipa15_1.csv")
```

```python
x = ["15","16","17","18","19"]
y = [i+1 for i in range(34)]
z = ["ipa","ips","smk"]
a = {}
k = 0
for i in x:
    for j in y:
        for k in z:
            key = "{}{}_{}".format(k,i,j)     
            a[key] = pd.read_csv(("prov_2/{}_prov/{}.csv").format(k,key)) 
```

```python
for i in x:
    for j in y:
        for k in z:
            key = "{}{}_{}".format(k,i,j)     
            a[key].to_json("../JSON/sekolah/{}{}_{}.json".format(k,i,j) , orient="records")
```

```python
b = {}
k = ["gab", "negri", "swas"]
for i in z:
    for j in k:
        key = "{}_{}".format(i,j)
        print(key)
        b[key] = pd.read_csv(("kab/{}.csv").format(key)) 
```

```python
k = ["gab", "negri", "swas"]
for i in z:
    for j in k:
        key = "{}_{}".format(i,j)
        x = b[key]
        x = x[~(x.JumlahSekolah ==0)]
        x.to_json("kab/jsonn/{}.json".format(key) , orient="records")
```
